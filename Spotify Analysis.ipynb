```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv("tracks.csv")
df_tracks.head()
```

```python
#null value
pd.isnull(df_tracks).sum()
```

```python
df_tracks.isna().sum()
```

```python
#total number of row and columns and check data type
df_tracks.info()
```

```python
#least popular song on spotify
sorted_df=df_tracks.sort_values('popularity',ascending=True)
sorted_df.head(10)
```

```python
df_tracks.describe().T
```

```python
#most popular song
most_popular=df_tracks.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
#most=df_tracks.sort_values('popularity',ascending=False)
#most[:10]
```

```python
#release date set as index
df_tracks.set_index('release_date',inplace=True)
```

```python
df_tracks.index=pd.to_datetime(df_tracks.index)
```

```python
df_tracks[['artists']].iloc[18]
```

```python
#convert duration into second
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x:round(x/1000))
df_tracks.drop('duration_ms',inplace=True,axis=1)
```

```python
corr_df=df_tracks.drop(['key','mode','explicit'],axis=1).corr()
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1f',cmap='inferno',linewidths=4)
heatmap.set_title('correlation between varable')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#creating sample data using 4% data
sample_df=df_tracks.sample(int(.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
#regression plot
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',line_kws={'color':'g'}).set(title='Loudness vs Energy correlation')
```

```python
sns.regplot(data=sample_df,x='acousticness',y='energy',scatter_kws={"color":"red"},line_kws={"color": "green"}).set(title='Acousticness vs Energy correlation')
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
years
```

```python
plt.figure(figsize=(25,20))
sns.displot(years,bins=20,discrete=True,aspect=2,kind='hist').set(title='Number of songs per year')
```

```python
total_dr=df_tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
#plt.figure(figsize=(20,15))
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title='year vs duration')
plt.xticks(rotation=90)
```

# Spotify Features Dataset Analysis

Duration of songs in different Genres

```python
df_genre=pd.read_csv('SpotifyFeatures.csv')
```

```python
df_genre.head()
```

```python
#duration of the songs for different genre
plt.title('Duration of the songs for different genre')
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel('Duration in mili second')
plt.ylabel('Genre')
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=df_genre.sort_values("popularity",ascending=False).head(10)
sns.barplot(y="genre",x="popularity",data=famous).set(title="top 10 genres by popularity")
```
